# DataFrame Pandas Singlestore Ibis

Variables d'environnement

```python
from dotenv import load_dotenv
import os

# Load environment variables from a .env file
load_dotenv()

# Access the variables
HOST_DML = os.getenv("HOST_DML")
USERDB = os.getenv("USERDB")
PASSDB = os.getenv("PASSDB")
```

## Connection à singlestore-ibis

Installer avec `pip install singlestoredb[dataframe]`

```python
import ibis
import pandas as pd


pd.options.display.max_colwidth = 120
ibis.options.verbose = False
ibis.options.interactive = True

conn = ibis.singlestoredb.connect(host=HOST_DML, port='3306', user=USERDB,
                  password=PASSDB, database='tpch')
```

Afficher les tables

```python
conn.show.tables()
```

Afficher un échantillon

```python
line = conn.table('lineitem')
line.head()
```

```python
orders = conn.table('orders')
orders.head()
```

Jointure

```python
merged_df = orders.join(line, [orders.o_orderkey == line.l_orderkey], how='inner')
merged_df.head()
```
